# Visualizations: 

```python
import pandas as pd
import altair as alt
bank = pd.read_csv("https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv")
bank_smp_1 = bank[bank["loan"] == "yes"]
bank_smp_1 = bank_smp_1.head(4999)
bank_smp_1_ages = bank_smp_1[["age", "loan"]]
alt.Chart(bank_smp_1_ages).mark_boxplot().encode(
  alt.X("age", title="Approved"),
  alt.Y("count(age)", title="Number of Approvals", scale = alt.Scale(domain=[0,10]))
  )

# print(bank_smp_1[["age", "loan"]])
```

# Model Creation 
### First Attempt 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

scaler = MinMaxScaler()

# reshaped = bank["age"].values.reshape(-1, 1)

# bank["age_scaled"] = scaler.fit_transform(reshaped)
# print(bank["age_scaled"])
bank_limited = bank[["age", "job", "marital"]]

# print(bank_limited)

X = pd.get_dummies(bank_limited, columns=["job", "marital"])
X["age"] = bank_limited["age"]

print(X)

Y = bank["loan"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)

clf = RandomForestClassifier()

clf.fit(X_train, Y_train)

clf.score(X_test, Y_test)
```

# Finding the best combination Test: 

```python
features = ["job", "loan", "marital", "education", "default", "housing", "contact",
            "month", "day_of_week", "poutcome"]
continuous = ["age", "campaign", "pdays", "previous", "emp.var.rate", "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed"]
X = pd.get_dummies(bank[features], columns= features)
X[continuous] = bank[continuous]
y = bank["y"]

print(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

clf = RandomForestClassifier()

clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```
